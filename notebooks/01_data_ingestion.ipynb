```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt


fund_master = pd.read_csv("../data/raw/fund_master.csv")

fund_master.columns = fund_master.columns.str.strip()

print("="*50)
print("DATASET SHAPE")
print("="*50)
print(fund_master.shape)

print("\n" + "="*50)
print("COLUMN NAMES")
print("="*50)
print(fund_master.columns.tolist())

print("\n" + "="*50)
print("DATA TYPES")
print("="*50)
print(fund_master.dtypes)

print("\n" + "="*50)
print("FIRST 5 ROWS")
print("="*50)
display(fund_master.head())

# MISSING VALUES

print("\n" + "="*50)
print("MISSING VALUES")
print("="*50)
display(fund_master.isnull().sum())

# DUPLICATE RECORDS

print("\n" + "="*50)
print("DUPLICATE RECORDS")
print("="*50)
print("Number of Duplicate Rows:", fund_master.duplicated().sum())

# AMC ANALYSIS

print("\n" + "="*50)
print("AMC ANALYSIS")
print("="*50)

print("Number of Unique AMCs:")
print(fund_master["amc_name"].nunique())

print("\nAMC Names:")
for amc in sorted(fund_master["amc_name"].unique()):
    print(amc)

print("\nAMC Scheme Counts:")
display(fund_master["amc_name"].value_counts().reset_index())

# CATEGORY ANALYSIS

print("\n" + "="*50)
print("CATEGORY ANALYSIS")
print("="*50)

print("Unique Categories:")
display(pd.DataFrame(
    sorted(fund_master["category"].dropna().unique()),
    columns=["Category"]
))

print("\nCategory Counts:")
display(fund_master["category"].value_counts())

# SUB CATEGORY ANALYSIS

print("\n" + "="*50)
print("SUB CATEGORY ANALYSIS")
print("="*50)

print("Unique Sub Categories:")
display(pd.DataFrame(
    sorted(fund_master["sub_category"].dropna().unique()),
    columns=["Sub Category"]
))

print("\nSub Category Counts:")
display(fund_master["sub_category"].value_counts())

# RISK GRADE ANALYSIS

print("\n" + "="*50)
print("RISK GRADE ANALYSIS")
print("="*50)

print("Unique Risk Grades:")
print(fund_master["risk_grade"].unique())

print("\nRisk Grade Counts:")
display(fund_master["risk_grade"].value_counts())

# FUND MANAGER ANALYSIS

print("\n" + "="*50)
print("FUND MANAGER ANALYSIS")
print("="*50)

print("Number of Fund Managers:")
print(fund_master["fund_manager"].nunique())

display(
    fund_master["fund_manager"]
    .value_counts()
    .head(10)
)

# EXPENSE RATIO ANALYSIS

print("\n" + "="*50)
print("EXPENSE RATIO ANALYSIS")
print("="*50)

expense_cols = [
    "expense_ratio_direct",
    "expense_ratio_regular"
]

display(
    fund_master[expense_cols].describe()
)

# VISUALIZATION - CATEGORY DISTRIBUTION

plt.figure(figsize=(10,5))

fund_master["category"].value_counts().plot(kind="bar")

plt.title("Category Distribution")
plt.xlabel("Category")
plt.ylabel("Count")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()

# VISUALIZATION - RISK GRADE DISTRIBUTION

plt.figure(figsize=(6,4))

fund_master["risk_grade"].value_counts().plot(kind="bar")

plt.title("Risk Grade Distribution")
plt.xlabel("Risk Grade")
plt.ylabel("Count")
plt.tight_layout()
plt.show()

# DATA SUMMARY

print("\n" + "="*50)
print("FINAL SUMMARY")
print("="*50)

print(f"Total Schemes      : {len(fund_master)}")
print(f"Unique AMCs        : {fund_master['amc_name'].nunique()}")
print(f"Unique Categories  : {fund_master['category'].nunique()}")
print(f"Unique SubCategory : {fund_master['sub_category'].nunique()}")
print(f"Risk Grades        : {fund_master['risk_grade'].nunique()}")
print(f"Fund Managers      : {fund_master['fund_manager'].nunique()}")
```
